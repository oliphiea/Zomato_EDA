```python
df_country=pd.read_excel(r'C:\Users\DELL\OneDrive\Documents\Zomatodataset\Country-Code.xlsx')
df_country.head()
```

```python
#joining two dataframes using countrycode in order to insert country column from second df 
df=pd.merge(df,df_country,on='Country Code',how='left')
df.head()
```

```python
df.dtypes
```

```python
df.Country.value_counts()
```

```python
country_names=df.Country.value_counts().index
country_values=df.Country.value_counts().values
```

```python
#pie chart
plt.pie(country_values,labels=country_names);  #which country has max orders 
```

```python
#top 3 countries that uses zomato using pie chart
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%');  
```

**Observation:** Zomato maximum transactions are from **India** after that **United States** and then **United Kingdom**

```python
df.columns
```

```python
ratings=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

**Observation:** 

1. When rating is between 4.5 to 4.9, the food was excellent
2. When rating is between 4.0 to 4.4, the food was very good
3. When rating is between 3.5 to 3.9, the food was good
4. When rating is between 2.5 to 3.4, the food was average
5. When rating is between 1.8 to 2.4, the food was poor

```python
plt.rcParams['figure.figsize']=(12, 6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings);
```

**Observation:**

1. Not rated count is very high
2. Maximum number of rating is between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green']);
```

```python
#Contries name that has given 0 rating
df[df['Rating text']=='Not rated'].groupby(['Country']).size().reset_index()
```

**Observation:** Maximum number of 0 ratings are from Indian customers

```python
#find out which currency is used by which country

df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```
